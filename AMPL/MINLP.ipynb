```python
from amplpy import AMPL, ampl_notebook
```

```python
SOLVER = 'knitro'

ampl = ampl_notebook(
    modules=['knitro'],  # modules to install
    license_uuid="------------Private License------------", )
```

```python
%%ampl_eval
reset;


var x >= 0;
var y >= 0;
var z integer >= 0, <= 15;

minimize obj:  x^2 + y^2 - 5 * z;

s.t. con1:  sqrt(x^2 + y^2) <= 10;
s.t. con2:  x + y >= 5;
```

```python
ampl.option["solver"] = SOLVER
ampl.solve()
```
